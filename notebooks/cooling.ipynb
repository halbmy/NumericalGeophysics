```python
import numpy as np 
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 100
```

```python
T0 = 1.0
tau = 0.015
dt = 0.01
t = np.arange(0.0, 1.001, dt)
Texp = np.ones_like(t) * T0
Timp = np.ones_like(t) * T0
Tmix = np.ones_like(t) * T0
for i, ti in enumerate(t[1:]):
    Texp[i+1] = Texp[i] * (1-dt/tau)
    Timp[i+1] = Timp[i] / (1+dt/tau)
    Tmix[i+1] = Tmix[i] * (1-dt/tau/2) / (1+dt/tau/2)
```

```python
fig, ax = plt.subplots()
tt = np.arange(0, 1.0001, 0.001)
ax.plot(tt, T0*np.exp(-tt/tau), "-", label="analytic")
ax.plot(t, Texp, "+--", label="explicit")
ax.plot(t, Timp, "x--", label="implicit")
ax.plot(t, Tmix, "v--", label="mixed")
ax.set_yscale('log')
ax.legend()
ax.grid()
```
