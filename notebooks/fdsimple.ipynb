## Finite difference modelling

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.arange(11)
print(len(x), x)
```

```python
A = np.zeros([len(x), len(x)])
for i in range(1, len(x)-1):
    A[i, i-1] = 1
    A[i, i] = -2
    A[i, i+1] = 1

A[0, 1] = 1
A[0, 0] = -2
A[-1, -1] = -1
A[-1, -2] = 1
```

```python
fig, ax = plt.subplots(figsize=(3, 2))
im = ax.matshow(A, vmin=-2, vmax=2, cmap="bwr");
plt.colorbar(im);
```

```python
f = np.zeros([len(x), 1])
f[0] = -10
f[len(x)//3] = -1
f[len(x)//3*2] = 2
f[-1] = +3
u = np.linalg.solve(A, f)
fig, ax = plt.subplots(figsize=(3, 2))
ax.plot(x, u);
ax.grid()
```

```python
plt.plot(A@u);
```
