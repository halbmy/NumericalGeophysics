```python
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 100
```

```python
import numpy as np 
import pygimli as pg 
```

```python
x = np.arange(4.)
x[-1] += 0.4
x[1] += 0.2
y = np.arange(3.)
y[-1] += 0.2
grid = pg.createGrid(x, y)
ax, _ = pg.show(grid)
ax.plot(pg.x(grid), pg.y(grid), "ko", markersize=6)
for i in range(len(x)-1):
    for j in range(len(y)-1):
        ij = f"{i},{j}"
        ax.text((x[i]+x[i+1])/2,
                (y[j]+y[j+1])/2,
                r"$\sigma_{"+ij+"}$")
ax.figure.savefig("grid32sigma.svg")
```

```python
ax.plot(x[:3], [1, 1, 1], "bo-", ms=12, lw=4)
ax.plot([1, 1, 1], y[:3], "bo-", ms=12, lw=4)
ax.plot(x[1], x[1], "ro", ms=12)
ax.figure.savefig("grid32sigmaStencil.svg")
ax.figure
```
