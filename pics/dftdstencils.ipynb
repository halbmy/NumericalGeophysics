```python
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 100
```

```python
import numpy as np 
import pygimli as pg 
import pygimli.meshtools as mt
```

```python
grid = mt.createGrid(5, 4)
```

```python
ax, cb = pg.show(grid)
ax.set_yticks([])
ax.set_xticks(np.arange(5))
ax.plot([1, 2, 3], [1, 1, 1], "bo-", ms=12, lw=4)
ax.plot([2, 2], [1, 2], "g-", lw=4)
ax.plot(2, 2, "ro", ms=12);
ax.figure.savefig("dtforward.svg")
```

```python
grid = mt.createGrid(5, 4)
ax, cb = pg.show(grid)
ax.set_yticks([])
ax.set_xticks(np.arange(5))
ax.plot([1, 2, 3], [2, 2, 2], "bo-", ms=12, lw=4)
ax.plot([2, 2], [1, 2], "go-", ms=12, lw=4)
ax.plot(2, 2, "ro", ms=12)
ax.figure.savefig("dtbackward.svg")
```

```python
ax, cb = pg.show(grid)
ax.set_yticks([])
ax.set_xticks(np.arange(5))
ax.plot([1, 2, 3], [1, 1, 1], "bo-", ms=12, lw=4)
ax.plot([1, 2, 3], [2, 2, 2], "bo-", ms=12, lw=4)
ax.plot([2, 2], [1, 2], "g-", ms=12, lw=4)
ax.plot(2, 2, "ro", ms=12)
ax.figure.savefig("dtmixed.svg")
```

```python
ax, cb = pg.show(grid)
ax.set_yticks([])
ax.set_xticks(np.arange(5))
ax.plot([1, 2, 3], [1, 1, 1], "bo-", ms=12, lw=4)
ax.plot([2, 2], [0, 2], "g-", lw=4)
ax.plot(2, 2, "ro", ms=12);
ax.plot(2, 0, "go", ms=12);
ax.plot(2, 1, "go", ms=8);
ax.figure.savefig("dtsecond.svg")
```
