```python
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 100
import numpy as np
import pygimli as pg
```

```python
x = np.arange(4.)
x[-1] += 0.4
x[1] += 0.2
y = np.arange(3.)
y[-1] += 0.2
grid = pg.createGrid(x, y)
ax, _ = pg.show(grid)
ax.plot(pg.x(grid), pg.y(grid), "ko", markersize=6)
for i in range(len(x)-1):
    for j in range(len(y)-1):
        ij = f"{i},{j}"
        ax.text((x[i]+x[i+1])/2,
                (y[j]+y[j+1])/2,
                r"$a_{"+ij+"}$", fontsize=14)
ax.figure.savefig("grid32a.svg")
```

```python
ax.plot(x[:3], [y[1]]*3, "bo-", ms=12, lw=4)
ax.plot([x[1]]*3, y[:3], "bo-", ms=12, lw=4)
ax.plot(x[1], y[1], "ro", ms=12)
ax.figure.savefig("grid32stencil.svg")
ax.figure
```
