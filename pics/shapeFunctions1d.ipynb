```python
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 100
```

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.array([0, 1., 2.1, 3.5, 5.0])
fig, ax = plt.subplots(figsize=(8, 4), nrows=2, sharex=True)
for i in range(len(x)):
    si = np.zeros_like(x)
    si[i] = 1
    ax[0].plot(x, si)
    dsdx = np.diff(si)/np.diff(x)
    ax[1].stairs(dsdx, x, ls="--", zorder=10, label=f"i={i}")
    # ax[1].stairs(np.hstack([dsdx[0], dsdx, dsdx[-1]]), 
                #  np.hstack([x[0], x, x[-1]]), ls="--")

ax[0].set_xticks(x)
ax[0].set_yticks([0, 1])
ax[1].set_yticks([-1, 0, 1])
ax[0].grid()
ax[1].grid()
ax[1].legend()
ax[1].set_xlabel("x")
ax[0].set_ylabel(r"$v_i$(x)")
ax[1].set_ylabel(r"$\partial v_i/\partial x$");
fig.tight_layout()
fig.savefig("shapefunctions1d.svg")
```
